**Задача 4.** *Интервальные оценки для параметра биномиального закона*

Используя выборку, сгенерированную вами в задаче 2 и считая параметр p неизвестным (k-дано), постройте для уровней доверия $1-\alpha=0.9, 0.95$ и $0.98$ симметричные интервальные оценки Клоппера-Пирсона ($\underline{p}$, $\overline{p}$) для вероятности успеха в одном испытании p.

Для тех же уровней найдите по ЦПТ приближенные доверительные интервалы для $p$.

Сравните полученные результаты и убедитесь, что полученные интервалы содержат истинное значение параметра.

Для одного из значений $\alpha$ постройте совмещенные графики функций распределения биномиальных законов B(k,p), B(k, $\underline{p}$), B(k,$\overline{p}$).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from scipy.stats import norm, beta, binom
from scipy.optimize import root_scalar
import scipy.stats as stats
```

```python
n = 180
k = 15
print('Вариант 8')
print('Объем выборки:', n)
print('Количество испытаний:', k)
```

Загружу выборку, сгенерированную в задаче 2

```python
X = np.array(pd.read_csv('task4.csv', header=None, sep=' ')[1], dtype=np.float64)
print('X:\n', X)
```

Задам статистику $K(\overrightarrow{X_n}) = \sum\limits_{i=1}^n X_i$

```python
K = X.sum()
print('Статистика K:', K)
```

По ЦПТ определим функцию $T(\theta)=\frac{K-nk\theta}{\sqrt{nk\theta (1-\theta)}} \thicksim N(0,1)$

```python
def T(teta):
    return (K - n * k * teta) / sqrt(n * k * teta * (1 - teta))
```

```python
xlist = np.linspace(0.001, 0.999, 1000)
Tlist = np.array([T(i) for i in xlist])
invnorm = np.array([norm.ppf(0.95, loc=0, scale=1) for i in xlist])
```

Оценю графически параметр p

```python
plt.figure(figsize=(15,7))
plt.subplot(1, 2, 1)
plt.plot(xlist, Tlist, label=r'$T(\theta)$')
plt.plot(xlist, invnorm, label='qnorm(0.95,0,1)')
plt.plot(xlist, -invnorm, label='- qnorm(0.95,0,1)')

plt.xlim(0.0, 1.0)
plt.ylim(-200.0, 200.0)
plt.xlabel(r'$\theta$')
plt.grid()
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xlist, Tlist, label=r'$T(\theta)$')
plt.plot(xlist, invnorm, label='qnorm(0.95,0,1)')
plt.plot(xlist, -invnorm, label='- qnorm(0.95,0,1)')

plt.title(r'Приближенная картина на $\theta \in [0.7; 0.9]$')
plt.xlim(0.7, 0.9)
plt.ylim(-10.0, 10.0)
plt.xlabel(r'$\theta$')
plt.grid()
plt.legend()

plt.show()
```

Пересечение построенной по ЦПТ функции T с обратными к нормальному закону $N(0,1)$ функциями, дадут границы приближенного интервала. Найдем координаты этих пересечений для $\theta$

```python
def lowup_interval(x, alpha, sign):
    if sign == '-':
        return T(x) - norm.ppf(1 - alpha / 2.0, loc=0, scale=1)
    else:
        return T(x) + norm.ppf(1 - alpha / 2.0, loc=0, scale=1)

def get_CLT_intervals(alpha):
    p_low = root_scalar(lowup_interval, args=(alpha, '-'), bracket=[0.6, 0.9])
    p_up = root_scalar(lowup_interval, args=(alpha, '+'), bracket=[0.6, 0.9])
    print(u'Приближенный интервал для α = {}: ({}, {})'.format(alpha, round(p_low.root, 3), round(p_up.root, 3)))

get_CLT_intervals(0.1)
get_CLT_intervals(0.05)
get_CLT_intervals(0.02)
```

Построение доверительных интервалов Клоппера-Пирсона строится на основе статистики $K(\overrightarrow{X_n}) = \sum\limits_{i=1}^n X_i$, теоретическая функция распределения которой, биномиальная, а значит имеет вид:
$$F(x,p)=\begin{cases}
0, x \leq 0
\\
\sum\limits_{j<x} C_{nk}^j\cdot p^j\cdot (1-p)^{nk-j}
\end{cases}$$
Далее найдем такие $C_1=C_1(p,\alpha)$ и $C_2=C_2(p,\alpha)$, что
$$\sum\limits_{j\leq C_1} C_{nk}^j\cdot p^j\cdot (1-p)^{nk-j} = P(K(\overrightarrow{X_n})\leq C_1)\leq \frac{\alpha}{2};$$
$$\sum\limits_{j > C_2} C_{nk}^j\cdot p^j\cdot (1-p)^{nk-j} = P(K(\overrightarrow{X_n}) < C_2)\geq 1 - \frac{\alpha}{2},$$
где $C_1$ и $C_2$ являются квантилями уровней $\frac{\alpha}{2}$ и $1-\frac{\alpha}{2}$ для биномиального закона $B(nk,p)$.

Тогда $P(C_1<K(\overrightarrow{X_n})<C_2)\geq 1-\alpha$. 

Строя оценку для $p$ множество $\{C_1(p,\alpha)<K(\overrightarrow{X_n})<C_2(p,\alpha)\}$ следует превратить в множество $\{\underline{p}(K(\overrightarrow{X_n}))<p<\overline{p}(K(\overrightarrow{X_n}))\}$, что приведет к уравнениям Клоппера-Пирсона для $\underline{p}$ и $\overline{p}$:
$$\sum\limits_{j=0}^{K(\overrightarrow{X_n})} C_{nk}^j\cdot \overline{p}^j\cdot (1-\overline{p})^{nk-j} = \frac{\alpha}{2};$$
$$\sum\limits_{j=0}^{K(\overrightarrow{X_n})-1} C_{nk}^j\cdot \underline{p}^j\cdot (1-\underline{p})^{nk-j} = 1-\frac{\alpha}{2}.$$

Последние уравнения решаются с помощью неполной бета-функции:
$$B_x(m,k)=\frac{1}{B(m,k)}\int\limits_0^x u^{m-1}(1-u)^{k-1}du, 0<x<1$$
которая связана с биномиальным законом следующим образом:
$$B_x(m+1,k-m)=1-\sum\limits_{j=0}^m C_{k}^j\cdot x^j\cdot (1-x)^{k-j}$$
С учетом этого уравнения Клоппера-Пирсона можно переписать в виде:
$$B_{\underline{p}} \Bigr(\sum\limits_{k=1}^n X_{k}, nk-\sum\limits_{k=1}^n X_{k}+1 \Bigl)=\frac{\alpha}{2}$$
$$B_{\overline{p}} \Bigr(\sum\limits_{k=1}^n X_{k} + 1, nk-\sum\limits_{k=1}^n X_{k} \Bigl)=1-\frac{\alpha}{2}$$

Используя встроенную функцию scipy.stats.beta.ppf, обратную к $B_x(m,k)$ получим границы доверительного интервала Клоппера-Пирсона

```python
def get_KP_intervals(alpha):
    p_low = beta.ppf(alpha / 2, K, n * k - K + 1)
    p_up = beta.ppf(1 - alpha / 2, K + 1, n * k - K)
    print('Доверительный интервал для α = {}: ({}, {})'.format(alpha, round(p_low, 3), round(p_up, 3)))

get_KP_intervals(0.1)
get_KP_intervals(0.05)
get_KP_intervals(0.02)
```

Исходя из полученных интервалов можно удостоверится, что во-первых, доверительные интервалы Клоппера-Пирсона и приближенные по ЦПТ интервалы схожи, и во-вторых, истинное значение $p = 0.8$ действительно содержится в них

Построю для $\alpha=0.05$ совмещенные графики функций распределения биномиальных законов B(k,p), B(k, $\underline{p}$), B(k,$\overline{p}$).

```python
p_low = 0.776
p_up = 0.807
p = 0.8

x = np.linspace(0, k + 1, k + 1)
cdf = binom.cdf

plt.figure(figsize=(14, 8))

plt.step(x, cdf(x, k, p_low), linestyle='-.', label='Биномиальный закон для p={}'.format(p_low))
plt.step(x, cdf(x, k, p), linestyle='solid', label='Биномиальный закон для p={}'.format(p))
plt.step(x, cdf(x, k, p_up), linestyle=':', linewidth=4, label='Биномиальный закон для p={}'.format(p_up))
plt.plot(x, [0.05 for i in x], linestyle='--', label='0.05')
plt.plot(x, [0.95 for i in x], linestyle='--', label='0.95')

plt.xlabel('$z$')
plt.title(r'Совмещенные графики функций распределения биномиальных законов для $\alpha=0.05$')
plt.legend()
plt.grid()
plt.show()
```

**Вывод**: в качестве оценки параметра $p$ подходят оба способа построения интервальной оценки: по методу Клоппера-Пирсона и по ЦПТ. Интервалы схожи и содержат истинное значение $p$ для уровней доверия $1-\alpha=0.9, 0.95$ и $0.98$
