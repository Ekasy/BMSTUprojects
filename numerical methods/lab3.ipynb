```python
def root_search(f, a, b, e, k=1):
    x = (a + b) / 2.0
    print('Итерация {}: a = {} b = {}\nx = {}\nошибка: {}'.format(k, a, b, x, abs(7.996589913 - x)))
    
    if f(x) * f(a) < 0:
        b = x
    elif f(b) * f(x) < 0:
        a = x
    
    if abs(b - a) > 2 * e:
        return root_search(f, a, b, e, k+1)
    
    return (a + b) / 2

left=5.0
right=10.0
epsilon=0.0001
y = lambda x: 13.190*x**3 - 211.210*x**2 - 844.8*x + 13516.790
x = root_search(y, left, right, epsilon)
print('x =', x)
```
