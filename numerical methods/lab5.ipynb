```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# отрезок
a = -1.5
b = 1.5

def f(x, N=8, n=52):
    return (10 + 0.5*N) / (1 + (20 + 0.25*N) * (1 + 0.05 * (53 - n)) * x**2)
```

```python
xlist = np.linspace(a, b, 1000)
flist = np.array([f(r) for r in xlist])

plt.figure(figsize=(12, 8))
plt.plot(xlist, flist)

plt.xlabel(r'x')
plt.ylabel(r'$f(x)$')
plt.grid()
plt.show()
```

Для равномерной сетки с 31 узлом

```python
avector = np.linspace(a, b, 31)
print('Равномерная сетка\n[', ', '.join([str(round(i, 1)) for i in avector]), ']')
```

```python
X = np.zeros((31, 31))

for i in range(31):
    for j in range(31):
        X[j][i] = avector[j]**i

pd.DataFrame(X)
```

```python
y = np.array([f(x) for x in avector])
y
```

```python
from numpy.linalg import solve

acoef = solve(X, y)
acoef[::-1]
```

```python
ylist = np.array([np.polyval(acoef[::-1], x) for x in np.linspace(a, b, 61)])

plt.figure(figsize=(12, 8))
plt.plot(np.linspace(a, b, 61), ylist)

plt.xlabel(r'x')
plt.ylabel(r'$f(x)$')

plt.grid()
plt.show()
```

```python
ylist = np.array([np.polyval(acoef[::-1], x) for x in np.linspace(a, b, 61)])

plt.figure(figsize=(12, 8))
plt.plot(xlist, flist, linestyle='--', linewidth=4, label=r'Исходная функция $f(\tau)$')
plt.plot(np.linspace(a, b, 61), ylist, linewidth=1, label=r'Интерполяционный полином Лагранжа')

plt.xlabel(r'$\tau$')
plt.ylabel(r'$f$')

plt.xlim(-1.6, 1.6)
plt.ylim(-1., 15)

plt.legend()
plt.grid()
plt.show()
```

Для чебышёвской сетки с 31 узлом

```python
from math import cos

def tau(j, k=30, a=-1.5, b=1.5):
    return (a + b) / 2 - (b - a) / 2 * cos((2*j + 1) * np.pi / (2 * (k + 1)))

tauvector = np.array([tau(j) for j in range(31)])
print('Чебышевская сетка\n[', ', '.join([str(round(i, 4)) for i in tauvector]), ']')
```

```python
X = np.zeros((31, 31))

for i in range(31):
    for j in range(31):
        X[j][i] = tauvector[j]**i

pd.DataFrame(X)
```

```python
y = np.array([f(x) for x in tauvector])
y
```

```python
acoef = solve(X, y)
acoef[::-1]
```

```python
ylist = np.array([np.polyval(acoef[::-1], x) for x in np.linspace(a, b, 61)])

plt.figure(figsize=(12, 8))
plt.plot(np.linspace(a, b, 61), ylist)

plt.xlabel(r'$\tau$')
plt.ylabel(r'$f$')

plt.grid()
plt.show()
```

```python
ylist = np.array([np.polyval(acoef[::-1], x) for x in np.linspace(a, b, 61)])

plt.figure(figsize=(12, 8))
plt.plot(np.linspace(a, b, 61), ylist, linewidth=1, label=r'Интерполяционный полином Лагранжа')
plt.plot(xlist, flist, linestyle='--', linewidth=2, label=r'Исходная функция $f(\tau)$')

plt.xlabel(r'$\tau$')
plt.ylabel(r'$f$')

plt.legend()
plt.grid()
plt.show()
```
